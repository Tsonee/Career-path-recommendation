```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn import preprocessing
```

```python
data = pd.read_excel("career_dataset_large.xlsx")
```

```python
data.head()
```

```python
data.describe()
```

```python
def bar_categorical(data, label, title = None, xlabel = 'Classes', ylabel='Counts'):
    if title is None: title = label
    plt.bar(data[label].value_counts().index, data[label].value_counts().values)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation='vertical')
    plt.show()
```

```python
bar_categorical(data, 'Education Level')
```

```python
data['Specialization'].value_counts().index
```

```python
bar_categorical(data, 'Specialization')
```

```python
bar_categorical(data, 'Certifications')
```

```python
plt.hist(data['CGPA/Percentage'])
plt.show()
```

```python
bar_categorical(data, 'Recommended Career')
```

```python
skills_for_plot = {}
for skill in data['Skills']:
    for s in skill.split(', '):
        if s in skills_for_plot:
            skills_for_plot[s]+=1
        else:
            skills_for_plot[s]=1
```

```python
plt.bar(skills_for_plot.keys(), skills_for_plot.values())
plt.title('Skills')
plt.xticks(rotation='vertical')
plt.xlabel('Classes')
plt.ylabel('Counts')
plt.show()
```

```python
columns_to_binarize = [ 0, 1, 2, 3]
binary_to_index = {}

for column_index in columns_to_binarize:
    for value in data.iloc[:, column_index]:
        if pd.notna(value):
            for splits in str(value).split(', '):
                binary_to_index[splits] = column_index
binary_to_index
```

```python
feature_names = list(binary_to_index.keys()) + ['CGPA/Percentage']
feature_names
```

```python
target = 'Recommended Career'
```

```python
binarized_attributes = pd.DataFrame(np.zeros( (len(data), len(feature_names)), dtype=int), columns = feature_names)
```

```python
for index in range(len(feature_names)):
    feature_name = feature_names[index]
    if feature_name not in binary_to_index:
        binarized_attributes[feature_name] = data[feature_name]
        continue
        
    column_index = binary_to_index[feature_name]

    for row in range(len(data)):
        value = data.iloc[row, column_index]
        if pd.notna(value):
            split = value.split(', ')
            binarized_attributes.iloc[row, index] = int(feature_name in split)

binarized_attributes
```

```python
data = binarized_attributes.join(data[target])
data.head()
```

```python
features = data.drop(target, axis=1)
target_binarized = data[target]

label_encoder= preprocessing.LabelEncoder()
target_encoded = label_encoder.fit_transform( data[target] )

target_encoded
```

```python
correlation_matrix = features.corr()
correlation_matrix
```

```python
number_of_features = len(feature_names)

plt.figure(figsize=(10, 10))
plt.xticks(range(number_of_features), feature_names, rotation='vertical')
plt.yticks(range(number_of_features), feature_names)
plt.imshow(correlation_matrix, cmap='ocean')
plt.colorbar()
plt.show()
```

```python
coefs = []

for feature_name in feature_names:
    column = features[feature_name]
    coefs.append(np.corrcoef(column, target_encoded)[0][1])

coefs
```

```python
dummies = pd.get_dummies(data[target]).astype('int64')
dummies
```

```python
corrcoefs_for_plot = pd.Series(coefs, index=feature_names)
corrcoefs_for_plot.sort_values(inplace=True)

corrcoefs_for_plot.plot(kind='barh', color='orange')
plt.title('Korelacija sa ciljnom promenljivom Recommended career')
plt.xlabel('Vrednost Pirsonovog koeficijenta korelacije')
plt.ylabel('Naziv atributa')
plt.show()
```
